## 环境相关
- 环境包含常用工具，用户可自行安装所需pip或npm包；若所需包大量缺失，该考虑换镜像而不是自行安装
- **用户自行安装的包会随容器重建而丢失**，在jupyter工作目录下或在jupyter文件中维护一个包安装文档或脚本是良好的习惯。
- 可用镜像有：
    - jupyter/scipy-notebook:3.1.1, 包含常用数据处理、机器学习工具包
    - transformers_notebook, 包含Tensorflow, Keras, Jax and PyTorch等AI工具包
    - 一个用户只能有一个容器，因此切换镜像时需要重建容器，为了保证安全性只有管理员才能杀掉旧容器，因此若需要切换镜像，需联系管理员去杀掉旧容器，才能按新镜像新建新容器
- 自行安装pip或npm包，建议用国内mirror:
    - 安装pip包可如：pip install numpy -i https://pypi.tuna.tsinghua.edu.cn/simple
    - 安装npm包可如：npm install -g pnpm --registry=https://registry.npm.taobao.org
